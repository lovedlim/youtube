### 반복문
- 반복해서 하는 작업수행
- 몇 번을 반복할지 고정/가변적일 때

```
subscriber_name
```

```
def printName(s):
  return print(s+':)')
```

```
for s in subscriber_name:
  printName(s)
```

```
for s in subscriber_name:
  print('퇴근후딴짓의 멤버: ',s)
```

```
def printCh3(s):
    print('채널명: ', c_name)
    print('구독자: ', s )
    print('----------------')
```

```
for s in subscriber_name:
  printCh3(s)
```
